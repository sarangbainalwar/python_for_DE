nested class - a class defined within another class 
    class outer:
        class inner:
group classes logically related.encapsulation

```python
class company:
    class employee:
        def __init__(self,name,position):
            self.name=name
            self.position=position
            
        def get_details(self):
            return f"{self.name} {self.position}"
        
    def __init__(self,company_name):
        self.company_name=company_name
        self.employees=[]
        
    def add_employee(self,name,position):
        new_employee=self.employee(name,position)
        self.employees.append(new_employee)
        
    def list_employees(self):
        return [emp.get_details() for emp in self.employees]
```

```python
company=company("krusty crab")
print(company.company_name)
```

```python
company.add_employee("eugne","manager")
company.add_employee("sarang","CEO")
company.add_employee("Saran","ASE")
```

```python
for employee in company.list_employees():
    print(employee)
```
