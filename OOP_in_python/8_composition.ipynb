aggregation : A relationship where one object contains references to other independent objects "has-a" relationship

composition : the composed object directly owns its components, which cannot exists independently "owns-a" relationship

```python
class engine:
    def __init__(self,hp):
        self.hp=hp

class wheel:
    def __init__(self,size):
        self.size=size
        
class car:
    def __init__(self,make,model,hp,size):
        self.make=make
        self.model=model
        self.engine=engine(hp)
        self.wheel=[wheel(size) for w in range(4)]
        
    def display_car(self):
        return f"{self.make} {self.model} {self.engine.hp} {self.wheel[0].size}"
```

```python
car1=car("ford","mustang",500,18)
car2=car("vw","gt",5000,125)
```
