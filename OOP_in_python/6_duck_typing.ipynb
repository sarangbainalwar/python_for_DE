Duck typing is another way to achieve polymorphism besides inheritance
object must have the minimum necessary attributes/methods
"If it looks like a duck and quack like a duck, it must be a duck"

```python
class animal:
    alive=True

class dog(animal):
    def speak(self):
        print("Wook")
        
class cat(animal):
    def speak(self):
        print("meow")
```

```python
class car:
    alive=False
    def speak(self):
        print("honk")
```

```python
animals=[dog(),cat(),car()]

for animal in animals:
    animal.speak()
    print(animal.alive)
```
