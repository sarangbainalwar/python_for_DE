```python
from abc import ABC, abstractmethod
```

```python
class vehical(ABC):
    
    @abstractmethod
    def go(self):
        pass
    
    @abstractmethod
    def stop(self):
        pass
```

```python
class car(vehical):
    def go(self):
        print(f"you drive the car")
        
    def stop(self):
        print(f"you stop the car")
```

```python
v1=car()
v1.go()
v1.stop()
```

```python
class motorcycle(vehical):
    def go(self):
        print(f"gooo")
        
    def stop(self):
        print(f"stooop")
```
